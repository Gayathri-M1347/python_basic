```python
import pandas as pd
import numpy as np

exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(exam_data , index=labels)
print("Select specific columns and rows:")
print(df.iloc[[1, 3, 5, 6], [0, 1, 3]])
```

###### Bar plot

```python
Murder = [13.2,10.0,8.1,8.8,9.0,7.9,3.3,5.9,15.4,17.4]

plt.boxplot(Murder)
plt.yticks(range(1, 22))
plt.ylabel("Murder")
plt.show()
```

###### Box Plot

```python
UrbanPop = ["58","48","80","50","91","78","77","72","80","60"]
Rape = [21.2,44.5,31,19.5,40.6,38.7,11.1,15.8,31.9,25.8]

# Generating the y positions. Later, we'll use them to replace them with labels.
y_positions = range(len(UrbanPop))

# Creating our bar plot
plt.bar(y_positions, Rape)
plt.xticks(y_positions, UrbanPop)
plt.ylabel("Rape (%)")
plt.title("Crime")
plt.show()
```

###### Pie chart

```python
labels = ['Civil', 'Electrical', 'Mechanical', 'Chemical']
sizes = [15, 50, 45, 10]


fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('Engineering Diciplines')


plt.show()

```
